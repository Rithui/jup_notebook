```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sin wave")
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up asubplot grid the has height 2 and width 1,
#and set the first such subplot
plt.subplot(2,1,1)

plt.plot(x,y_sin,'b--')
plt.title("sin wave")
print("\n")
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title("cos wave")
```

```python
#bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y,align='edge')
plt.bar(x2,y2,color='g',align='center')
plt.title("Bar Graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
