```python
import seaborn as sns
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.corr()
```

```python
sns.jointplot(x='petal_width',y='sepal_length',data=iris,kind='reg')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(data=iris,hue='petal_length')
```

```python
sns.distplot(iris['sepal_length'],kde=True,bins=10)
```

```python
sns.countplot(x='species',data=iris)
```

```python
sns.barplot(y='species',x='petal_length',data=iris)
```

```python
sns.boxplot(x='petal_width',y='sepal_length',data=iris)
```

```python
sns.violinplot(x='species',y='sepal_width',data=iris)
```
